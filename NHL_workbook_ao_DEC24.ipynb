The idea that PDO regresses to 100 (or 1.000) stems from the law of averages and the way shooting percentage (SH%) and save percentage (SV%) typically behave over the long run. Here's why 100 (or 1.000) is significant:

1. Total Shots = Goals + Saves
In hockey, the total number of shots taken by both teams combined should roughly equal the total number of shots faced by goalies, because every shot either results in a goal or is saved. Therefore, the sum of a team's shooting percentage (SH%) and the opponent's save percentage (SV%) is, in theory, balanced over time.

Mathematically, when you add up shooting and save percentages, it should always equal around 100% (or 1.000 when expressed as a decimal).

Shooting percentage (SH%) is the percentage of shots that a team converts into goals.
Save percentage (SV%) is the percentage of shots that a goalie stops.
So, if a team has a high shooting percentage (i.e., they are scoring on a higher-than-average number of shots), it's likely that their opponents are having a lower save percentage, meaning the opposing goalie is letting more goals in than usual. Similarly, if a team has an unusually high save percentage, it might mean their shooting percentage is lower than normal, and they are getting fewer goals despite creating chances.

2. The Law of Averages (Regression to the Mean)
Over a large number of games, most teams will experience a balanced distribution between SH% and SV%, meaning that extremes in either stat (very high or very low) are typically unsustainable in the long run. Here’s why:

Shooting percentage (SH%) is influenced by a variety of factors, including player skill, shot selection, goaltender performance, and random chance.
Save percentage (SV%) is also influenced by random chance and the goalie’s ability to stop pucks, but it will often even out over time as goalies face more shots.
When a team has a PDO significantly above 100, it typically indicates that either their SH% is unusually high (they’re scoring more goals than expected from their shots) or their SV% is unusually high (the goalie is stopping more shots than expected). While this can happen over a short period, it's unlikely to remain sustainable over an entire season.

Conversely, a team with a PDO below 100 likely has a low shooting percentage or a poor save percentage, or both, suggesting that they are not finishing their scoring chances effectively, or their goalie is struggling. Over time, this team may improve as their shooting and save percentages revert to more typical levels.

3. PDO as a Measure of "Luck"
PDO has become a shorthand for measuring a team's "luck" because it reflects how much a team's performance is based on things that can fluctuate (like shooting and saving percentages). In the short term, a team can experience "good luck" (high PDO) or "bad luck" (low PDO), but in the long term, the regression to 100 suggests that these high or low numbers will tend to normalize as the season progresses.

4. Statistical Balance
When you sum SH% and SV%, the total should roughly equal 100% over the course of a season, because:

A high SH% means the team is converting more shots into goals, but this often coincides with a lower SV% by the opposing goalie (more goals against).
A high SV% means the goalie is stopping more shots, but this often coincides with a lower SH% (because fewer goals are being scored).
Therefore, 100 is considered the baseline, or the average, because it reflects a balance between offensive and defensive performance. The league-wide average PDO will hover around 100 over time, as the law of averages dictates that teams with unusually high PDOs (either due to high shooting or save percentages) will see their numbers regress, and teams with low PDOs will experience improvement as their shooting and/or goaltending stabilize.

Example:
Let’s say a team has a PDO of 105 (or 1.050), meaning their shooting and save percentages combined are unusually high. This might suggest that the team is getting "lucky" in the short term, with an efficient offense and a goalie who is making exceptional saves. Over time, this PDO will likely regress to 100, and the team might see their shooting percentage and/or save percentage drop to more typical levels.

Conversely, if a team has a PDO of 95 (or 0.950), they may be experiencing bad luck, with low shooting efficiency or poor goaltending. Over time, their PDO is expected to regress upwards toward 100, meaning they should eventually start scoring more goals and/or getting better goaltending.

In Summary:
PDO regresses to 100 (or 1.000) because, over a large number of games, the sum of a team's shooting and save percentages should be close to 100%.
100 is significant because it represents an average or normal balance between a team’s ability to score and its ability to prevent goals.
Teams with PDO significantly higher or lower than 100 are often experiencing unsustainable extremes that will likely regress toward 100 over time.
This concept helps analysts and coaches distinguish between a team that is performing well due to skill and a team that is benefiting from good or bad luck.

Significant NHL analytics per ChatGPT: Summary of Key NHL Analytics:
Corsi & Fenwick – Shot attempt metrics used to measure puck possession.
Corsi/Fenwick For % (CF% / FF%) – Percentage of shot attempts for a team/player.
Expected Goals (xG) – Measures the quality of scoring chances.
High-Danger Scoring Chances (HDSC) – Tracks high-quality scoring opportunities.
Zone Starts & Zone Exits – Measures shifts starting in offensive/defensive zones and ability to move the puck out of the defensive zone.
Relativity Metrics (e.g., Corsi Rel%) – Measures a player’s impact on their team's possession.
Expected Goals Against (xGA) – Tracks quality of chances allowed.
Save Percentage (SV%) – Goalie effectiveness at stopping shots.
Shooting Percentage (SH%) – Team/player shooting efficiency.
Game Score (GS) – A holistic measure of player performance.
Time on Ice (TOI) – A measure of a player’s involvement in the game.
These analytics provide a much deeper understanding of team and player performance beyond traditional stats like goals, assists, and plus/minus, allowing analysts to assess the true underlying factors that contribute to success in hockey.
