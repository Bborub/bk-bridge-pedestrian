```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Download stock price data
start = "2023-01-01"
end = "2023-08-25"
data = yf.download("SPY", start=start, end=end)

# Perform exponential smoothing
alpha = 0.2  # Smoothing factor (0 < alpha < 1)
smoothed_data = pd.Series(data["Close"]).ewm(alpha=alpha).mean()

# Plot the stock price data and exponential smoothing
plt.figure(figsize=(12, 6))
plt.plot(data.index, data["Close"], label="Stock Price")
plt.plot(smoothed_data.index, smoothed_data, label="Exponential Smoothing")
plt.title("Stock Price Data with Exponential Smoothing")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the stock price data
start = "2020-01-01"
end = "2023-08-25"
data = yf.download("SPY", start, end)

# Calculate the rolling mean and standard deviation
window = 20  # Adjust this window size as per your preference
rolling_mean = data["Close"].rolling(window).mean()
rolling_std = data["Close"].rolling(window).std()

# Calculate the z-scores for each data point
z_scores = (data["Close"] - rolling_mean) / rolling_std

# Define a threshold for anomaly detection
threshold = 2.5  # Adjust this threshold as per your preference

# Find the anomalies
anomalies = data[np.abs(z_scores) > threshold]

# Set style using seaborn
sns.set(style="darkgrid", palette="Set2")

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the stock price data
sns.lineplot(data=data["Close"], ax=ax, label="Stock Price")

# Highlight the anomalies
sns.scatterplot(data=anomalies["Close"], ax=ax, color="red", label="Anomalies")

# Set titles and labels
ax.set_title("Stock Price Data with Anomalies")
ax.set_xlabel("Date")
ax.set_ylabel("Price")

# Format x-axis tick labels
ax.xaxis.set_major_locator(plt.MaxNLocator(10))
ax.xaxis.set_major_formatter(plt.DateFormatter("%Y-%m-%d"))

# Rotate x-axis tick labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.metrics import f1_score

# Load the stock price data
start = "2022-08-25"
end = "2023-08-25"
data = yf.download("SPY", start, end)

# Prepare the data for anomaly detection
df = data[["Close"]].copy()
df.reset_index(inplace=True)
df.rename(columns={"Date": "ds", "Close": "y"}, inplace=True)

# Split the dataset into training and validation sets
train_size = int(len(df) * 0.8)  # Adjust the split ratio as per your preference
train_data = df.iloc[:train_size]
val_data = df.iloc[train_size:]

# Define a range of contamination values
contamination_values = np.arange(0.01, 0.11, 0.01)  # Adjust the range and step size as per your preference

# Perform cross-validation
best_f1 = 0.0
best_contamination = None

for contamination in contamination_values:
    # Convert the stock price data to a numpy array
    X_train = train_data["y"].values.reshape(-1, 1)

    # Detect anomalies using the Isolation Forest algorithm
    model = IsolationForest(contamination=contamination)
    model.fit(X_train)

    # Predict anomalies on the validation set
    X_val = val_data["y"].values.reshape(-1, 1)
    y_pred = model.predict(X_val)

    # Evaluate performance using F1-score
    f1 = f1_score(np.ones_like(y_pred), -y_pred)  # Assuming all anomalies are labeled as 1
    if f1 > best_f1:
        best_f1 = f1
        best_contamination = contamination

# Print the optimal contamination value
print("Optimal Contamination Value:", best_contamination)
```

The z-score tells us how many standard deviations away a data point is from the mean. A positive z-score indicates that the data point is above the mean, while a negative z-score indicates that it is below the mean.

By setting a threshold, you can identify outliers or anomalies based on the magnitude of the z-score. If a data point has a z-score that exceeds the threshold (e.g., a z-score greater than 2 or 3), it is considered an anomaly.

Using z-scores for anomaly detection in a stock chart involves calculating the z-scores for each data point based on a rolling mean and standard deviation. This rolling approach accounts for the changing characteristics of the data over time. Data points with z-scores exceeding a predefined threshold are then flagged as anomalies.

In the previous code example, we calculated the z-scores for the stock price data using a rolling mean and standard deviation. We then used a threshold value (e.g., threshold = 2.5) to identify anomalies by comparing the absolute value of the z-scores to the threshold.

Please note that the choice of threshold depends on your specific application and the desired sensitivity to anomalies. Adjusting the threshold allows you to control the strictness of anomaly detection.
